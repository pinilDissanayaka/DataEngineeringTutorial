```python
from mysql import connector
import pandas as pd
```

```python
try:
    con=connector.connect(
                          host="localhost",
                          user="root",
                          password=""
                )
except connector.Error as e:
    print(e)
```

```python
def createDB(connector, database):
    try:
        cur=connector.cursor()
        cur.execute(f'''CREATE DATABASE IF NOT EXISTS {database}''')
    except:
        print("Couldn't connect cursor.")
    finally:
        cur.close()
```

```python
def createTable(connector, query):
    try:
        cur=connector.cursor()
        cur.execute(query)
    except:
        print("Couldn't connect cursor.")
    finally:
        cur.close()
```

```python
try:
    con=connector.connect(
                          host="localhost",
                          user="root",
                          password="",
                        database="Country"
                )
except connector.Error as e:
    print(e)
```

```python
accountsCountryQuery='''CREATE TABLE IF NOT EXISTS AcountsCountry(
                            Code VARCHAR(10) PRIMARY KEY,
                            LongName VARCHAR(20),
                            CurrencyUnit VARCHAR(20),
                            TableName VARCHAR(20));
                            '''
```

```python
createTable(connector=con, query=accountsCountryQuery)
```

```python
accountsCountryInset='''INSERT INTO AcountsCountry VALUES
                        (%s, %s, %s, %s);
'''
```

```python
try:
    cur=con.cursor()
    for _, row in accountsCountry.iterrows():
        cur.execute(accountsCountryInset, list(row))
except:
    print("Couldn't insert values.")
finally:
    cur.close()
```
